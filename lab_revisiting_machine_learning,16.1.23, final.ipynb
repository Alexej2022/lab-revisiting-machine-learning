```python
print(categoricals['GENDER'].value_counts())
categoricals['GENDER'] = categoricals['GENDER'].fillna('F')
```

```python
def gender_transform(x):
    if  x in ['M', 'F']:
        return x
    else:
        return 'other'
```

```python
categoricals['GENDER'] = list(map(gender_transform, categoricals['GENDER']))
```
